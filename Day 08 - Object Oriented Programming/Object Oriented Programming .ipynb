````python
class Possesion:
    # Class attribute
    # this variable will be shared among all the methods
    num_possesions = 0 # Global or Shared Attribute

    # python's way of understanding, what you want to take into the function -- or what inputs/variables
    # __init__ : it is a Constructor --- helps us build objects later on
    # everytime we create an object, this function is executed
    def __init__(self, kind, age, color, brand): # this is the place to put any parameters
        # self here targets the object that we will create later
        # save the data for an object
        self.kind = kind # attributes of the class
        self.age = age
        self.color = f"this is "
        self.brand = brand
        Possesion.num_possesions = Possesion.num_possesions + 1
        Possesion.num_possesions += 1

    # variable inside a method --- we are not using any attributes from objects here
    def greet():
        lucky_number = 42 # local to this method
        print(f"{lucky_number} says Hello!!! and Possesions are currently {Possesion.num_possesions}")

    # access attributs of an object inside a method
    # we are using attributes from objects here
    def details(self):
        # this should give most important details of an object
        # new_number = 23
        print(f"Type is: {self.kind}\nThe age is: {self.age}\nThe brand is: {self.brand}")
    
    # Getter setter methods! 
````
