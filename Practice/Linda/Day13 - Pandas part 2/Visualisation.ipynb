Diff types of EDA in Data Visualisation part:
1. Univariate - analysis of only one variable
2. Bivariate - of 2 variables
3. Multivariate - of more than 2 variables
- does not only mean for visualisation

EDA can be:
1. Quantitative - descriptive - summary tables, central tendencies (mean, median, mode), dispersion, variance
2. Graphical - Visual - Bar Chart, histogram, scatter plot, box plot etc

```python
import matplotlib.pyplot as plt
import seaborn as sns

categories = [1,2,3,4,5]
values = [100,20,100,40,80]

bar_color = ["red", "pink", "orange", "yellow", "green"]

# hashvalues are used for colors in web
# hexa decimal is a number system - 0 1 2 3 4 5 6 7 8 0 A B C D E F
bar_color1 = ["#fc5757", "#b83535", "#460808", "#1B0101", "#3d0808"]
#bar_color1 = #ffffff

hatch = ['\\\\', '|||', '+','O', 'X']

#bar chart
plt.bar(categories, values, 
        color = bar_color1, # color argument decides the colors at bars
        edgecolor = "black",   #"black" or #ffffff
        linewidth = 2,
        alpha = 0.7, # transperancy
        hatch = "\\"  #"\\"
        ) 

# title
plt.title("Bar Chart For Show Only")

# label for x axis
plt.xlabel("categories")

# label for y axis
plt.ylabel("values")

plt.show()
```

```python
# exercise:
## import employees dataset
## create a bar graph for salaries of different employees
## use pandas for the data parts

import pandas as pd 
import matplotlib.pyplot as plt

df_emp = pd.read_csv("employees.csv")
#df_emp

# Extract columns from DataFrame
Name = df_emp["Name"]
Salary = df_emp["Salary"]

# title
plt.title("Employees Information")

# label for x axis
plt.xlabel("Employee Name")

# label for y axis
plt.ylabel("Salary")

bar_color = ["red", "pink", "orange", "yellow", "green"]
#bar_color1 = #ffffff

#bar chart
plt.bar(
    Name, 
    Salary, 
    color = bar_color, # color argument decides the colors at bars
    edgecolor = "grey",   #"black" or #ffffff
    linewidth = 2,
    alpha = 0.7, # transperancy
    hatch = "\\"  #"\\"
) 

plt.show()
```

### Histogram
- Univariate analysis
- is generally used to check frequency --- Frequency on y-axis by default
- can be used for number ranges too
- not just categories
- histogram can be continous bars as well

```python
import numpy as np

np.random.seed(42)
data = np.random.normal (size = 1000) # 1000 data points
# data.min()

#plt.hist(data)
plt.hist(data, bins = 20, edgecolor = "#4b188d")

#sms.histplot(data, bins = 20, edgecolor = "#2e973f")

plt.title("Normal Distribution") # title
plt.xlabel("Data") # label X axis
plt.show()
```

### line plot
- Bivariate
- also has two axis
- and there should two data arrays

- 1 point - infinite lines
- 2 points - single line only


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([x for x in range(100)])

y = x * 2

plt.plot(x,y)
```

```python
x = np.array([x for x in range(2)])

x = np.linspace(0,7,100)
y = x ** 2

y = np.sin(x)

y2 = np.cos(x)

plt.plot(x,y, color ="blue", label = "sine")

plt.plot(x,y2, color ="red", label = "cosine")

plt.legend()

plt.grid()

plt.plot(x,y)
```
